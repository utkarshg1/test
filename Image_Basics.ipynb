# How images are stored in a computer

```python
from warnings import filterwarnings
filterwarnings("ignore")
```

# Load the image

```python
import cv2

# By default image is read as BGR
img = cv2.imread("/content/CuteCat.jpeg")
img
```

```python
# Converting BGR to RGB
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img2
```

```python
img2.shape
```

```python
import matplotlib.pyplot as plt

plt.imshow(img2)
plt.show()
```

# Splitting Red, green, blue channels from the image

```python
r, g, b = cv2.split(img2)
```

```python
r
```

```python
g
```

```python
b
```

```python
plt.imshow(r, cmap="Reds")
plt.show()
```

```python
plt.imshow(g, cmap="Greens")
plt.show()
```

```python
plt.imshow(b, cmap="Blues")
plt.show()
```

# Plot the orignal image with channels

```python
plt.figure(figsize=(12, 8))

plt.subplot(1, 4, 1)
plt.imshow(img2)
plt.title("Orignal Image")
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(r, "Reds")
plt.title("Red Channel")
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(g, "Greens")
plt.title("Green Channel")
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow(b, "Blues")
plt.title("Blue Channel")
plt.axis('off')

plt.show()
```
